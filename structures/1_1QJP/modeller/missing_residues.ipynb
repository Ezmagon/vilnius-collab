```python
from modeller.automodel import *
```

```python
env = environ()
```

```python
env.io.atom_files_directory = ['.', '../atom_files']
```

```python
class MyModel(automodel):
    def select_atoms(self):
        return selection(self.residue_range('1', '21'),
                         self.residue_range('39', '51'),
                         self.residue_range('85', '91'),
                         self.residue_range('160','165'))
```

```python
a = MyModel(env, alnfile= 'fill.ali',
           knowns = '1qjp', sequence = '1qjp_fill')
```

```python
a.starting_model = 1
a.ending_model = 1
```

```python
a.make()
```
